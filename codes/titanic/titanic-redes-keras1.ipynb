¡Claro! El siguiente código en Python utiliza una red neuronal para predecir los supervivientes del Titanic. Utilizaremos la librería de aprendizaje automático Keras para construir la red neuronal y la base de datos Titanic para entrenar el modelo.

```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Cargamos la base de datos Titanic
df = pd.read_csv("./data/Titanic.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Explore a pattern
```

```python
women = df.loc[df.Sex == 'female']["Survived"]
rate_women = sum(women)/len(women)

print("% of women who survived:", rate_women)
```

```python
men = df.loc[df.Sex == 'male']["Survived"]
rate_men = sum(men)/len(men)

print("% of men who survived:", rate_men)
```

```python
# Eliminamos columnas no relevantes para el análisis
df = df.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"])
```

```python
df
```

```python
# Convertimos valores textuales a numéricos
df["Sex"] = pd.factorize(df["Sex"])[0]
df["Embarked"] = pd.factorize(df["Embarked"])[0]
```

```python
df.info()
```

```python
# Rellenamos valores faltantes con la media
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
# Separamos los datos en conjuntos de entrenamiento y prueba
train_data = df.sample(frac=0.8, random_state=0)
test_data = df.drop(train_data.index)
```

```python
train_data.info()
```

```python
test_data.info()
```

```python
# Definimos los datos de entrada y salida
train_x = train_data.drop(columns=["Survived"])
train_y = train_data["Survived"]
test_x = test_data.drop(columns=["Survived"])
test_y = test_data["Survived"]
```

```python
# Definimos la red neuronal
model = Sequential()
model.add(Dense(12, input_dim=7, activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
```

```python
# Compilamos el modelo
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Entrenamos el modelo
model.fit(train_x, train_y, epochs=150, batch_size=10)
```

```python
# Evaluamos el modelo
_, accuracy = model.evaluate(test_x, test_y)
print("Accuracy: %.2f%%" % (accuracy * 100))
```

Este código carga la base de datos Titanic, realiza algunas transformaciones y limpieza de datos, y separa los datos en conjuntos de entrenamiento y prueba. Luego, define una red neuronal con una capa de entrada de 12 nodos, una capa oculta de 8 nodos y una capa de salida de 1 nodo. La función de activación utilizada en las capas ocultas es "relu" y en la capa de salida es "sigmoid". El modelo se compila con la función de pérdida "binary_crossentropy" y el optimizador "adam". Finalmente, el modelo se entrena durante 150 épocas y se evalúa su precisión en el conjunto de prueba. El resultado es impreso en pantalla.
